```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Blood Pressure (mmHg) (Normal: 130/80)

| Date/Time | 8-11 am | 9-11 pm |
|-----------|---------|---------|
| 5/6       | 180/92  | 165/96  |
| 6/6       | 164/92  | -       |
| 7/6       | 155/91  | -       |
| 8/6       | 164/108 | -       |


```python
# Convert the Markdown table to a pandas DataFrame
data = {
    'DateTime': ['2023-06-05 09:00', '2023-06-05 21:00', '2023-06-06 09:00', 
                 '2023-06-07 09:00', '2023-06-08 09:00'],
    'Systolic': [180, 165, 164, 155, 164],
    'Diastolic': [92, 96, 92, 91, 108]
}

df = pd.DataFrame(data)
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Create the plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")

# Plot readings
systolic_line, = plt.plot(df['DateTime'], df['Systolic'], 'b-', label='Systolic')
diastolic_line, = plt.plot(df['DateTime'], df['Diastolic'], 'r-', label='Diastolic')

# Add reference lines for normal blood pressure
plt.axhline(y=140, color='g', linestyle=':', label='Normal Systolic')
plt.axhline(y=90, color='g', linestyle=':', label='Normal Diastolic')

# Add data labels
for i, (datetime, systolic, diastolic) in enumerate(zip(df['DateTime'], df['Systolic'], df['Diastolic'])):
    plt.annotate(f'{systolic}', (datetime, systolic), textcoords="offset points", xytext=(0,10), ha='center')
    plt.annotate(f'{diastolic}', (datetime, diastolic), textcoords="offset points", xytext=(0,-15), ha='center')

plt.title('Blood Pressure Readings')
plt.xlabel('Date and Time')
plt.ylabel('Blood Pressure (mmHg)')
plt.legend()

# Format x-axis
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m %H:%M'))

# Show the plot
plt.tight_layout()
plt.show()
```

### Heart rate (bpm) (Normal: 60-100)

| Date/Time | Heart Rate |
|-----------|---------|
| 5/6       | 64      |
| 6/6       | 70      |
| 7/6       | 82      |
| 8/6       | 82      |


```python
# Convert the heart rate data to a pandas DataFrame
heart_rate_data = {
    'DateTime': ['2023-06-05', '2023-06-06', '2023-06-07', '2023-06-08'],
    'HeartRate': [64, 70, 82, 82]
}

df_hr = pd.DataFrame(heart_rate_data)
df_hr['DateTime'] = pd.to_datetime(df_hr['DateTime'])

# Create the plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")

# Plot heart rate readings
plt.plot(df_hr['DateTime'], df_hr['HeartRate'], 'b-o', label='Heart Rate')

# Add reference lines for normal heart rate range
plt.axhline(y=60, color='g', linestyle=':', label='Normal Range')
plt.axhline(y=100, color='g', linestyle=':')

# Color areas above and below normal range
plt.fill_between(df_hr['DateTime'], 100, 110, color='red', alpha=0.2, label='Above Range')
plt.fill_between(df_hr['DateTime'], 50, 60, color='blue', alpha=0.2, label='Below Range')

# Add data labels
for i, (datetime, hr) in enumerate(zip(df_hr['DateTime'], df_hr['HeartRate'])):
    plt.annotate(f'{hr}', (datetime, hr), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Heart Rate Readings')
plt.xlabel('Date')
plt.ylabel('Heart Rate (bpm)')
plt.legend()

# Format x-axis
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m'))

# Set y-axis range
plt.ylim(50, 110)

# Show the plot
plt.tight_layout()
plt.show()
```

### Temperature (C)

| Date/Time | Temperature |
|-----------|-------------|
| 5/6       | 36.4C      |
| 6/6       | 36.4C      |
| 7/6       | 36.4C      |
| 8/6       | 36.4C      |


```python
# Convert the temperature data to a pandas DataFrame
temperature_data = {
    'DateTime': ['2023-06-05', '2023-06-06', '2023-06-07', '2023-06-08'],
    'Temperature': [36.4, 36.4, 36.4, 36.4]
}

df_temp = pd.DataFrame(temperature_data)
df_temp['DateTime'] = pd.to_datetime(df_temp['DateTime'])

# Create the plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")

# Plot temperature readings
plt.plot(df_temp['DateTime'], df_temp['Temperature'], 'r-o', label='Temperature')

# Add reference lines for normal temperature range
plt.axhline(y=36.5, color='g', linestyle=':', label='Normal Range')
plt.axhline(y=37.5, color='g', linestyle=':')

# Add data labels
for i, (datetime, temp) in enumerate(zip(df_temp['DateTime'], df_temp['Temperature'])):
    plt.annotate(f'{temp}C', (datetime, temp), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Temperature Readings')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()

# Format x-axis
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m'))

# Set y-axis range
plt.ylim(35.5, 38.5)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Convert the Capillary glucose data to a pandas DataFrame
glucose_data = {
    'DateTime': ['2023-06-05', '2023-06-06', '2023-06-07', '2023-06-08'],
    'GlucoseLevel': [16.9, 8.2, 11.3, 7.9]
}

df_glucose = pd.DataFrame(glucose_data)
df_glucose['DateTime'] = pd.to_datetime(df_glucose['DateTime'])

# Create the plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")

# Plot glucose readings
plt.plot(df_glucose['DateTime'], df_glucose['GlucoseLevel'], 'g-o', label='Glucose Level')

# Add reference lines for normal glucose range (assuming 4.0 - 7.8 mmol/L as normal)
plt.axhline(y=4.0, color='b', linestyle=':', label='Normal Range')
plt.axhline(y=7.8, color='b', linestyle=':')

# Color areas above normal range
plt.fill_between(df_glucose['DateTime'], 7.5, 18, color='red', alpha=0.2, label='Above Range')

# Add data labels
for i, (datetime, glucose) in enumerate(zip(df_glucose['DateTime'], df_glucose['GlucoseLevel'])):
    plt.annotate(f'{glucose}', (datetime, glucose), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Capillary Glucose Readings')
plt.xlabel('Date')
plt.ylabel('Glucose Level (mmol/L)')
plt.legend()

# Format x-axis
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m'))

# Set y-axis range
plt.ylim(0, 18)

# Show the plot
plt.tight_layout()
plt.show()
```
